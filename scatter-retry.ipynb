```python
# Dependencies and Setup
# Import PyPlot, Pandas, and NumPy
import pandas as pd
import matplotlib.pyplot as plt
import numpy
```

```python
# Read in the csv and preview it
# It should be in '../Resources/cardata.csv'
horsepowerdf = pd.read_csv("../Resources/cardata.csv")
horsepowerdf 
```

```python
# Select only the columns that you need from the original data frame 
horsepowerDFcolumns = horsepowerdf[["car name","horsepower", "weight"]]
horsepowerDFcolumns = horsepowerDFcolumns.set_index(["car name"])
horsepowerDFcolumns['horsepower'] = pd.to_numeric(horsepowerDFcolumns['horsepower'], errors='coerce')

# Clean data issues
# horsepowerDFcolumns

# Create your plot
horsepowerDFcolumns.plot.scatter(x='weight', y='horsepower')
plt.show()
```
